```python
import itertools
```

```python
def ToListMatrix(hadamard_matrix):
    dim = len(hadamard_matrix[0])
    matrix = [[i for i in hadamard_matrix[j]] for j in range(0,dim)]
    for row in matrix:
        for i in range(dim):
            if row[i] == -1:
                row[i] = 0
    return matrix
```

```python
def Find_Allowed_Cosets(matrix, col1, col2):
    vec_len = len(matrix)/2
    list1 = []
    list2 = []
    i = 0
    for row in matrix:
        x = (row[col1] - row[col2]) % 2
        if i < vec_len:
            list1.append(x)
        else:
            list2.append(x)
        i += 1
    bad_cosets = []
    for i in range(len(list1)):
        for j in range(len(list2)):
            if list1[i] == list2[j]:
                x = (j - i) % vec_len
                bad_cosets.append(x)
    allowed_cosets = []
    for j in range(vec_len):
        if j not in bad_cosets:
            allowed_cosets.append(j)
    return allowed_cosets
```

```python
hadamard20 = ToListMatrix(hadamard_matrix(20))
hadamard20
```

```python
pairs = [(y,x) for (x,y) in itertools.combinations(range(20), 2)]
```

```python
coset_table20 = {}
for pair in pairs:
    cyc_classes = Find_Allowed_Cosets(hadamard20, pair[0], pair[1])
    coset_table20[pair] = cyc_classes
coset_table20
```
